```python
!pip install apyori
import numpy as np
import pandas as pd
from apyori import apriori
dataset = pd.read_csv("/content/drive/MyDrive/Market_Basket_Optimisation.csv")
#dataset = pd.read_csv("/content/drive/MyDrive/accidents.csv")
dataset
print(dataset.shape)
transactions = []
#for using aprori need to convert data in list format..
# transaction = [['apple','almonds'],['apple'],['banana','apple']]....
for i in range(0, 7500):
 transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
print(len(transactions))
rules = apriori(transactions = transactions, min_support = 0.001, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
# all rules need to be converted in a list
results = list(rules)
print("Rules:")
print(results)
#convert result in a dataframe for further operation
def inspect(results):
    lhs         =[tuple(result[2][0][0])[0] for result in results]
    rhs         =[tuple(result[2][0][1])[0] for result in results]
    supports    =[result[1] for result in results]
    confidences =[result[2][0][2] for result in results]
    lifts        =[result[2][0][3] for result in results]
    return list (zip(lhs, rhs, supports, confidences, lifts))
#making a table from a result dataframe
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ["Left hand side", "Right hand side", "Support", "Confidence", "Lift"])
print(resultsinDataFrame)
#sorting the table
resultsinDataFrame.nlargest(n = 10, columns = "Lift")
resultsinDataFrame.nlargest(n = 10, columns = "Confidence")
```
